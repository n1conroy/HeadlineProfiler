```python
# Function to remove tags
def remove_tags(html):
  
    # parse html content
    soup = BeautifulSoup(html, "html.parser")
  
    for data in soup(['style', 'script']):
        # Remove tags
        data.decompose()
  
    # return data by retrieving the tag content
    return ' '.join(soup.stripped_strings)
```

```python
import pandas as pd
from bs4 import BeautifulSoup
texts = []
sources = []
intents = []
pdObj = pd.read_json(filename, orient='index')
for i,p in enumerate(pdObj):
  if 'body' in pdObj[p]['results']['content']:
    text = pdObj[p]['results']['content']['body']
    sttext = remove_tags(text)
    texts.append(sttext)
    source = pdObj[p]['results']['source']
    sources.append(source)


```

```python
d = {"text": texts, "source": sources}
df = pd.DataFrame(d)
```
